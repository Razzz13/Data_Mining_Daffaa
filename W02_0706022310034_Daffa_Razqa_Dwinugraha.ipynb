```python
# 1. Load the dataset

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2. Inspect the dataset

url = "https://raw.githubusercontent.com/NathaliaMinoque/datasets/refs/heads/main/jabodetabek_house_price.csv "
df = pd.read_csv(url, sep=',')

df.head()
```

```python
df.info()
```

```python
df.describe()
```

Keyfinding:

dari dataset diatas, terdapat perbedaan harga antar lokasi. selain itu Harga per Meter Persegi Berbeda Signifikan antar Lokasi. selain itu terdapat missing value terutama di bagian building_orientation hingga 1647 data selain itu masih ada building_age dan year_built yang sama 1445 data

```python
# 3. Handle missing values

col = ["bedrooms","bathrooms","land_size_m2","building_size_m2","floors"]
for cols in col:
    df[cols] = df[cols].fillna(df[cols].median())

# Convert 'building_age' to numeric, coercing errors to NaN
df['building_age'] = pd.to_numeric(df['building_age'], errors='coerce')
# Fill remaining NaNs with the median after coercion
df['building_age'] = df['building_age'].fillna(df['building_age'].median())

kol = ["certificate","year_built","property_condition","building_orientation","furnishing"]
df[kol] = df[kol].fillna("Unknown")
df = df.dropna(subset=["ads_id","property_type"])
```

```python
df.info()
```

```python
df['price_bin_quartile'] = pd.qcut(df['price_in_rp'], q=4, labels=['Low','Medium-Low','Medium-High','High'])
```

```python
# Feature Engineering + EDA
bins = [0, 5, 10, 20, 50, df['building_age'].max()]
labels = ['0-4 years', '5-9 years', '10-19 years', '20-49 years', '50+ years']
df['building_age_category'] = pd.cut(df['building_age'], bins=bins, labels=labels, right=False, include_lowest=True)

# Analyze price by building age category

price_by_building_age_category = df.groupby('building_age_category')['price_in_rp'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=price_by_building_age_category.index, y=price_by_building_age_category.values)
plt.title('Average House Price by Building Age Category')
plt.xlabel('Building Age Category')
plt.ylabel('Average Price (in Rp)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Insight:

Bangunan yang berusia 10-19 tahun rata rata harganya lebih tinggi dibanding kan usia lain nya

```python
# Analyze price by number of bedrooms and bathrooms

plt.figure(figsize=(12, 6))
sns.scatterplot(x='bedrooms', y='price_in_rp', data=df, alpha=0.5)
plt.title('House Price vs. Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price (in Rp)')
plt.show()

plt.figure(figsize=(12, 6))
sns.scatterplot(x='bathrooms', y='price_in_rp', data=df, alpha=0.5)
plt.title('House Price vs. Number of Bathrooms')
plt.xlabel('Number of Bathrooms')
plt.ylabel('Price (in Rp)')
plt.show()
```

Insight:

harga rumah meningkat jika bertambahnya jumlah kamar tidur atau kamar mandi

```python
# Analyze price by furnishing

price_by_furnishing = df.groupby('furnishing')['price_in_rp'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=price_by_furnishing.index, y=price_by_furnishing.values)
plt.title('Average House Price by Furnishing')
plt.xlabel('Furnishing')
plt.ylabel('Average Price (in Rp)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Insight:

jika rumah sudah dalam kondisi furnished atau semi, rata rata harga akan lebih mahal


```python
# Analyze price by city

price_by_city = df.groupby('city')['price_in_rp'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=price_by_city.index, y=price_by_city.values)
plt.title('Average House Price by City')
plt.xlabel('City')
plt.ylabel('Average Price (in Rp)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Insight:

rata rata harga rumah di jakarta pusat lebih tinggi dibandingkan dengan harga rumah di daerah lain

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['price_in_rp'], kde=True, bins=50, log_scale=True)
plt.title('Distribution of House Prices (Log Scale)')
plt.xlabel('Price (in Rp - Log Scale)')
plt.ylabel('Frequency')
plt.show()
```
